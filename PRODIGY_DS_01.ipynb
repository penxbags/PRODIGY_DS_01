```python
#Data Visualization of a Catagorical Variable-origin using a bar graph

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the DataFrame from cars.csv dataset downloaded from kaggle
path="C:\\Users\\pattnaik\\Desktop\\Prodigy InfoTech Internship\\datasets\\cars_automobile.csv"
data = pd.read_csv(path)

df = pd.DataFrame(data)

# Count the occurrences of each category in the 'origin' column
origin_counts = df['origin'].value_counts()

# Plotting the bar chart using Seaborn

# Set the color palette to pastel for the plot
sns.set_palette("pastel")

# Create a figure with specified size
plt.figure(figsize=(6, 4))

# Plot the bar chart using Seaborn
sns.barplot(x=origin_counts.index, y=origin_counts.values)

# Add a title to the plot
plt.title('Distribution of Car Origins')

# Add a label to the x-axis
plt.xlabel('Origin')

# Add a label to the y-axis
plt.ylabel('Count')

# Display the plot
plt.show()

#CONCLUSION:count of cars from usa is highest or maximum.
```
